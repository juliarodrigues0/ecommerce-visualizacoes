```python
!pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/ecommerce_estatistica.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.histplot(df['Nota'], bins=10, kde=False)
plt.title("Distribuição das Notas")
plt.xlabel("Nota")
plt.ylabel("Frequência")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='N_Avaliações', y='Qtd_Vendidos_Cod', data=df)
plt.title("Avaliações vs Quantidade Vendida")
plt.xlabel("Número de Avaliações")
plt.ylabel("Quantidade Vendida")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Calor das Correlações")
plt.show()
```

```python
plt.figure(figsize=(8,4))
df.groupby('Marca')['Nota'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Marcas por Nota Média")
plt.xlabel("Marca")
plt.ylabel("Nota Média")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['Gênero'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Distribuição por Gênero")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(6,4))
precos = pd.to_numeric(df['Preço_MinMax'], errors='coerce').dropna()
sns.kdeplot(x=precos, fill=True)
plt.title("Distribuição de Densidade dos Preços Normalizados")
plt.xlabel("Preço Normalizado")
plt.ylabel("Densidade")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.regplot(x='Preço_MinMax', y='Qtd_Vendidos_Cod', data=df, scatter_kws={'alpha':0.5})
plt.title("Relação entre Preço e Quantidade Vendida")
plt.xlabel("Preço Normalizado")
plt.ylabel("Quantidade Vendida")
plt.show()
```
